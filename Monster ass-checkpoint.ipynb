# Loading json file into dataframe with help of pandas

```python
import pandas as pd

df = pd.read_json(r"D:\jupyter files\assign_chubb_engineering  (1).json")
df
```

# Transposing dataframe so that company name could be rows(index) so it can be easy to fetch data


```python
df11 = df.T
df11
```

# Resetting index and changing the name of previous index

```python
df11.reset_index(inplace=True)
df11.rename(columns={'index':'Comp name'},inplace =True)
df11
```

# making new dataframe(a) and using json_normalize function on employees col. of df11 to pick out specific keys and ignoring the keys which we don't wan't

#  using those keys as columns and renaming those columns according to us. 

```python

a = pd.json_normalize(df11['employees'])
a = a.loc[:,['count','freshers_count','from_iit.count','from_iit.freshers_count','from_nit.count','from_nit.freshers_count']]
a = pd.concat([a,df11['company_uuid'],df11['Comp name'],df11['is_score_calculated'],df11['Department_Tier-1_College_Density_Score'],df11['Department_Tier-1_College_Freshers_Density_Score']],axis=1)
a.rename(inplace=True,columns={'company_uuid':'Company uuid','Comp name':'Company Name','count':'Employees Count','freshers_count':'Fresher Count','from_iit.count':'IIT Count','from_iit.freshers_count': 'IIT Fresher Count','Department_Tier-1_College_Density_Score':'Tier-1 College Density Score','Department_Tier-1_College_Freshers_Density_Score':'Tier1 College Freshers Density Score','is_score_calculated':'Is score calculated','from_nit.count':'NIT Count','from_nit.freshers_count':'NIT Fresher Count'})
a
```

```python
cond = [(a['Employees Count']>=1)&(a['Employees Count']<=10),
        (a['Employees Count']>=11)&(a['Employees Count']<=50),
        (a['Employees Count']>=51)&(a['Employees Count']<=100),
        (a['Employees Count']>=101)&(a['Employees Count']<=250),
       (a['Employees Count']>=251)&(a['Employees Count']<=500),
       (a['Employees Count']>=510)&(a['Employees Count']<=1000),
       (a['Employees Count']>=1001)&(a['Employees Count']<=5000),
       (a['Employees Count']>=5001)&(a['Employees Count']<=10000),
       (a['Employees Count']>10000)]
choice = [1,2,3,4,5,6,7,8,9]
a['category number'] = np.select(cond,choice)
a
```
